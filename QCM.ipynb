### Classification with a SamplerQNN

```python
# construct QNN with the QNNCircuit's default ZZFeatureMap feature map and RealAmplitudes ansatz.
qc = QNNCircuit(num_qubits=feature_map.num_qubits, feature_map=feature_map, ansatz=ansatz)
estimator_qnn = EstimatorQNN(circuit=qc)
qc.draw(output="mpl")
```

```python
# callback function that draws a live plot when the .fit() method is called
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
# construct neural network classifier
estimator_classifier = NeuralNetworkClassifier(
    estimator_qnn, optimizer=COBYLA(maxiter=60), callback=callback_graph
)
```

```python
from sklearn.preprocessing import MinMaxScaler
features = MinMaxScaler().fit_transform(X)
```

```python
from qiskit.circuit.library import ZZFeatureMap

num_features = features.shape[1]

feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)
feature_map.decompose().draw(output="mpl", fold=20)
```

```python
from qiskit.circuit.library import RealAmplitudes

ansatz = RealAmplitudes(num_qubits=num_features, reps=3)
ansatz.decompose().draw(output="mpl", fold=20)
```

```python
from qiskit_algorithms.optimizers import COBYLA

optimizer = COBYLA(maxiter=100)
```

```python
from qiskit.primitives import Sampler

sampler = Sampler()
```

```python
from matplotlib import pyplot as plt
from IPython.display import clear_output

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)


def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
print(len(train_features))
print(len(train_labels))
```

```python
import time
from qiskit_machine_learning.algorithms.classifiers import VQC

vqc = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

start = time.time()
vqc.fit(train_features, train_labels)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

```python
# Evaluate the Quantum Machine Learning model on the test dataset
predictions = vqc.predict(test_features)

# Calculate accuracy
accuracy = np.sum(predictions == test_labels) / len(test_labels)

print(f"Quantum Machine Learning Model Accuracy on Test Dataset: {accuracy:.2%}")
```
